```python
# 1. Find the second largest element in an array without sorting.
list = [5,8,9,10,12,40,50,23]
for i in range(len(list)):
    for j in range(len(list)):
        if list[i]>list[j]:
            list[i],list[j] = list[j],list[i]
print(list[1])
```

```python
# pyramid

for i in range(4,-1,-1):
    print(' '*i,end='')
    for j in range(4,i-1,-1):
        print('* ',end='')
    print()
```

```python
n=9
for i in range(n):
    if i <= n // 2:
        num = i + 1
    else:
        num = n - i
    for j in range(n):
        if i==j or i+j == n-1:
            print(num, end="")
        else:
            print(' ', end='')
    print()
```
