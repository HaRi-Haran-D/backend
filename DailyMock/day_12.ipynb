```python
# 1. Find the second largest element in an array without sorting.
list = [5,8,9,10,12,40,50,23]
for i in range(len(list)):
    for j in range(len(list)):
        if list[i]>list[j]:
            list[i],list[j] = list[j],list[i]
print(list[1])
```

```python
# 3. Check if two strings are anagrams.
str1 = 'listen'
str2 = 'SILENT'

def anagram(str1,str2):
    s1=sorted(str1.lower().replace(' ',''))
    s2=sorted(str2.lower().replace(' ',''))
    return s1==s2
result=anagram(str1,str2)
print(f'Is "{str1}" an anagram of "{str2}"? {result}')
```

```python
def rotate(arr, k):
    n = len(arr)
    k = k % n

    def reverse(start, end):
        while start < end:
            arr[start], arr[end] = arr[end], arr[start]
            start += 1
            end -= 1
    # Step 1: Reverse whole array
    reverse(0, n - 1)
    # Step 2: Reverse first k elements
    reverse(0, k - 1)
    # Step 3: Reverse remaining elements
    reverse(k, n - 1)
    return arr

arr = [1,2,3,4,5]
print(rotate(arr, 2))
```
