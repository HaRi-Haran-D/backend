```python
# 1. Write a Python program to check for Pangram?

def is_pangram(text):
    text = text.lower()
    for char in "abcdefghijklmnopqrstuvwxyz":
        if char not in text:
            return False
    return True

sentence = "The quick brown fox jumps over the lazy dog"
#sentence = 'It is a Pangram'

if is_pangram(sentence):
    print("It is a Pangram")
else:
    print("Not a Pangram")

```

```python
# 2. Write a Python program to basic Data Structure Operations (e.g., list manipulation, string manipulation)?

# list manipulation
my_list = [10, 20, 30, 40]

# Add element
my_list.append(50)
print("After append:", my_list)

# Insert element
my_list.insert(1, 15)
print("After insert:", my_list)

# Remove element
my_list.remove(30)
print("After remove:", my_list)

# Pop element
my_list.pop()
print("After pop:", my_list)

# Update element
my_list[2] = 100
print("After update:", my_list)

# Traverse list
print("List elements:")
for item in my_list:
    print(item, end=" ")
print()


#string manipulation
my_string = "Python Programming"

# Length
print("Length:", len(my_string))

# Convert to upper case
print("Uppercase:", my_string.upper())

# Convert to lower case
print("Lowercase:", my_string.lower())

# Replace word
print("Replace:", my_string.replace("Python", "Java"))

# Slice string
print("Slice:", my_string[0:6])

# Check substring
if "Python" in my_string:
    print("Substring found")
else:
    print("Substring not found")
```

```python
# 3. Write a Python program to find Minimum Element in a List?

list = [3,45,12,6,7,9,1,100]
n=list[0]

# Minimum
for i in list:
    if i < n:
        n=i
print(n)

#Maximum
for i in list:
    if i > n:
        n=i
print(n)
```

```python
# 4. Write a Python program to Remove Duplicates from a String?

def duplicate(str):
    n = ''
    for i in str:
        if i not in n:
            n+=i
    return n
str = 'aaabbcbca'
print(duplicate(str))
```
