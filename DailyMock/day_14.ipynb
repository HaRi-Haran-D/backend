```python
# 1. Find the index of the maximum value in an array.

arr = [10,34,5,67,12]
largest = arr[0]
for i in range(1,len(arr)):
    if largest < arr[i]:
        largest = arr[i]
print(largest)
```

```python
# 2. Sort a NumPy array in ascending and descending order.
import numpy as np

n = np.array([1,2,3,4])

print(np.sort(n))
print(np.sort(n)[::-1])
print(np.max(n))
```

```python
s = "A man, a plan, a canal: Panama"
rev=''
ori=''
sym = [' ',':',',']

for i in range(len(s)):
    if s[i] in sym:
        continue
    elif s[i] == s[i].upper():
        ori+=s[i].lower()
    else:
        ori+=s[i]
print(ori)
for i in range(len(s)-1,-1,-1):
    if s[i] in sym:
        continue
    elif s[i] == s[i].upper():
        rev+=s[i].lower()
    else:
        rev+=s[i]
print(rev)
if rev != ori:
    print(False)
else:
    print(True)
```
