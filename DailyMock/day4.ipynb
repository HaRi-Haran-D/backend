```python
# 2. Find the missing number in a list of 1 to n.
n=[1,2,4,5,6,8,9,12]
nn=[]
for i in range(0,len(n)):
    if n[i+1] - n[i] == 1:
        nn+=[i]
    else:
        nn+=[n[i]+1]
print(nn)
```

```python
# 4. Reverse a list in-place without using [::-1] or reverse().
n=[1,2,4,5,6,8,9,12]
nn=[]
for i in range(len(n),0,-1):
    nn+=[n[i-1]]

print(nn)
```

```python
# 5. Find the first non-repeating element in a list.
list = [1,2,3,1,2,4,6,4]
n=[]
for i in list:
    if i not in n:
        n+=[i]
print(n)
```
