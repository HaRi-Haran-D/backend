```python
# 1. Find the largest element in a list
list = [1,2,3,4,7,9,11]
for i in range(len(list)):
    for j in range(0,len(list)):
        if list[i]>list[j]:
            list[i],list[j]=list[j],list[i]
print(list[0])
```

```python
# 2. Find the smallest element in a list
list = [11, 9, 7, 4, 3, 2, 1]

for i in range(len(list)):
    for j in range(0,len(list)):
        if list[i]<list[j]:
            list[i],list[j]=list[j],list[i]
print(list[0])
```
