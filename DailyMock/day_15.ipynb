```python
# 1. Reverse a string without using slicing ([::-1]).
# Input: "PythonRocks" → Output: "skcoRnohtyP

n = "PythonRocks"
rev = ''
for i in range(len(n)-1,-1,-1):
    rev += n[i]
print(rev)
```

```python
# 2. Check if a number is a palindrome.

# Input: 121 → Output: True

# Input: 123 → Output: False

def palin(n):
    o = n
    rev=0
    while o > 0:
        rem = o% 10
        rev = (rev * 10)+ rem
        o = o//10
    return rev == n
print(palin(121))
print(palin(123))
```

```python
# 3. Find the second largest element in a list.

# Input: [10, 35, 50, 23, 50] → Output: 50

n = [10, 35, 50, 23, 50]
for i in range(len(n)):
    for j in range(len(n)):
        if n[i] > n[j]:
            n[i],n[j] = n[j],n[i]
print(n[1])
```

```python
# 4. Count the frequency of each character in a string.

# Input: "programming" → Output: {'p':1,'r':2,'o':1,'g':2,'a':1,'m':2,'i':1,'n':1}

str = "programming"
count = {}

for i in str:
    if i in count:
        count[i]+=1
    else:
        count[i]=1

print(count)
```

```python
# 5. Merge two sorted lists into a single sorted list.

# Input: [1,3,5] and [2,4,6] → Output: [1,2,3,4,5,6]
a = [1,3,5]
b = [2,4,6]
c=a+b
for i in range(len(c)):
    for j in range(len(c)):
        if c[i] < c[j]:
            c[i],c[j] = c[j],c[i]
print(c)
```

```python
# 6. Find all pairs in a list whose sum equals a target value.

# Input: [1, 2, 3, 4, 5], target 5 → Output: [(1,4),(2,3)]

def sum(num,target):
    f=[]
    for i in range(len(num)):
        for j in range(i,len(num)):
            if num[i] + num[j] == target:
                f+=[(num[i],num[j])]
    return f
print(sum(num=[1, 2, 3, 4, 5],target=5))
```
