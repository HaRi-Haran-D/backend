```python
def insertion_sort(arr):
    n = len(arr)

    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

arr = [8, 3, 5, 2, 9, 1]
sorted_arr = insertion_sort(arr)
print(sorted_arr)
```

```python
# 5. input arr = {4,5,6,5,4,3,1,2,2,4}
#    output [4,4,4,2,2,5,5,1,3,6]

arr = [4,5,6,5,4,3,1,2,2,4]
freq = {}
for i in arr:
    freq[i] = freq.get(i, 0) + 1

result = sorted(arr, key=lambda x: (-freq[x], x))

print(result)
```
