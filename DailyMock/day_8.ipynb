```python
# 1. Write a recursive function to calculate Fibonacci numbers.
def fib(start,end):
    num1 = 0
    num2 = 1
    for i in range(start, end+1):
        result = num1 + num2
        print(num1, end=' ')
        num1 = num2
        num2 = result
fib(0,10)
```

```python
# 2. Write a recursive function to calculate factorial.
def fact(n):
    if n == 0:
        return 1
    else:
        return n * fact(n-1)
fact(6)
```

```python
# 3. Create a BankAccount class with methods: deposit(), withdraw(), check_balance().
class Account():
    def __init__(self, act_no, balance):
        self.act_no = act_no
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            return "Invalid amount"
        else:
            self.balance += amount
            return self.balance

    def withdraw(self, amount):
        if amount >= self.balance:
            return "Insufficient Balance"
        else:
            self.balance -= amount
            return self.balance

    def check_balance(self):
        return self.balance

n=Account('Hari',1000)
n.deposit(100)
n.withdraw(1000)
print(n.check_balance())
```
