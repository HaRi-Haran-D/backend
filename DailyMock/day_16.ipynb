```python
def subsets(nums):
    result = [[]]

    for num in nums:
        new = []
        for subset in result:
            new.append(subset + [num])
        result.extend(new)

    return result
print(subsets([1, 2, 3]))
```

```python
# 1. Reverse a string
str = 'string'

rev=''
for i in range(len(str)-1,-1,-1):
    rev+=str[i]
print(rev)
```

```python
# 2. Palindrome check

# str = 'madam'
str='str'
rev=''
for i in range(len(str)-1,-1,-1):
    rev+=str[i]
if str == rev:
    print(True)
else:
    print(False)
```

```python
# 3. Find largest number in list

arr = [1,2,3,4,10,6]
lar=arr[0]

for i in arr:
    if i>lar:
        lar=i
print(lar)
```
