```python
class Solution(object):
    def runningSum(self, nums):
        num=0
        n=[]
        while num <= len(nums):
            num+=1
            if num==0:
                n+=[nums[num]]
            else:
                n+=[nums[num] + nums[num-1]]
        return n

f=Solution()
print(f.runningSum(nums=[1,2,3,4]))
```
