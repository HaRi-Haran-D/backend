```python
# Pascal's Triangle
# Input: numRows = 5
# Output: [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]

class Solution(object):
    def generate(self, numRows):
        a=1
        for i in range(numRows,0,-1):
            print(' '*i, end="")
            for j in range(i-1,numRows):
                print('*', end=' ')

            print()
n=Solution()
n.generate(numRows=5)
```
