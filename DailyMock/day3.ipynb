```python
# 1. You have a list of numbers representing daily sales. How would you find the top 3 highest sales?

sales = [45,67,12,34,675,898,13]

for i in range(len(sales)):
    for j in range(i+1, len(sales)):
        if sales[i] < sales[j]:
            sales[i],sales[j] = sales[j],sales[i]
print(sales[:3])
```

```python
# 2. You receive a tuple of coordinates (x, y) multiple times. How would you count how many times each coordinate appears?

coordinates = [(2, 3),(3, 5),(2, 3),(5, 6),(3, 5),(2, 3),(2, 4),(2, 4),(2, 3)]
count = {}

for point in coordinates:
    if point in count:
        count[point] += 1
    else:
        count[point] = 1
print(count)
```

```python
# 3. You want to reverse a list of logs but keep the original list intact. How would you do it efficiently?
list = [45,67,12,34,675,898,13]
empt=[]

for i in range(len(list),0,-1):
    empt+=[list[i-1]]
print(empt)
```

```python
# 4. You have a list of strings containing duplicates. How would you create a list of unique strings while maintaining their original order?

strings = ["Hari", "JK", "JK", "Yuvi", "Hari"]
uni = []
seen = set()

for s in strings:
    if s not in seen:
        uni+=[s]
        seen.add(s)
print(uni)
```

```python
# 5. You have a list of employee names and a separate list of their salaries. How would you combine them into a dictionary?

name = ["Hari", "JK", "Yuvi"]
salary = [30000, 80000, 55000]

emp = dict(zip(name, salary))
print(emp)

#OR
emp = {}

for i in range(len(name)):
    emp[name[i]] = salary[i]
print(emp)
```

```python
# 6. You have two lists of user IDs from two systems. How would you find users present in both systems?

user1 = [101, 102, 103, 104]
user2 = [103, 104, 105, 106]
both=[]

for i in user1:
    if i in user2:
        both+=[i]
print(both)
```

```python
# 7. You have a list of emails and want to remove duplicates efficiently. Which collection would you use?

emails = ["fb@gmail.com","ins@gmail.com","fb@gmail.com","ins@gmail.com"]
email = []

for i in emails:
    if i not in email:
        email+=[i]
print(email)
```
