```python
# List Comprehension Mastery
# 1. Write a single-line list comprehension that takes a list of strings, filters out strings shorter than 4 characters, and converts the remaining strings to uppercase.

words = ["apple", "bat", "cherry", "dog", "elderberry"]
# Output: ['APPLE', 'CHERRY', 'ELDERBERRY']

def filter(words, length):
    list=[]
    for i in words:
        if len(i) > length:
            list+=[i.upper()]
    return list
print(filter(words,4))
```

```python
# Dictionary Merging with Logic
# 2. Write a function that merges two dictionaries. If a key exists in both dictionaries, sum their values. If a key exists in only one, include it as is.

dict_a = {'a': 10, 'b': 20}
dict_b = {'b': 5, 'c': 15}
# Output: Merged Dictionary: {'a': 10, 'b': 25, 'c': 15}

def merge(d1,d2):
    result = d1.copy()
    for key, value in d2.items():
        result[key] = result.get(key,0) + value
    return result
print(merge(dict_a,dict_b))
```

```python
# Frequency Map with Counter
# 3. Create a function that takes a string and returns a count of how many times each character appears. Ignore spaces and make it case-insensitive.

text = "Python Programming"
# # output: Character Frequency:
# Counter({'p': 2, 'y': 1, 't': 1, 'h': 1, 'o': 2, 'n': 2, 'r': 2, 'a': 1, 'g': 2, 'm': 2, 'i': 1})

from collections import Counter
def char(text):
    clean = text.lower().replace(" ","")
    return Counter(clean)
print(char(text))
```

```python
# Flatten a Nested List
# 5. Write a recursive function that takes a list containing other lists (of any depth) and returns a single “flat” list of all elements.

nested = [1, [2, 3], [4, [5, 6]], 7]
# Output: Flattened: [1, 2, 3, 4, 5, 6, 7]

def flatten(lst):
    flat_list = []

    for item in lst:
        if isinstance(item, list):
            # Recursion: if it's a list, flatten it and extend our results
            flat_list.extend(flatten(item))
        else:
            # Base case: it's a single value, just add it
            flat_list.append(item)
    return flat_list

nested_data = [1, [2, 3], [4, [5, 6]], 7]
result = flatten(nested_data)

print(f"Original:  {nested_data}")
print(f"Flattened: {result}")
```
