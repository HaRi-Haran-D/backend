```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

n = 20
primes = []
for num in range(2, n + 1):
    if is_prime(num):
        primes.append(num)
print(f'All prime numbers from 1 to {n}: {primes}')
```

```python
# Bubble Sort
list = [19,2,31,45,6,11,121,27]
def bubblesort(list):
    for i in range(len(list)):
        for j in range(len(list)):
            if list[i]<list[j]:
                list[i],list[j] = list[j],list[i]
    return list

print(bubblesort(list))
```

```python
n = 'madam'
def palin(n):
    left = 0
    right = len(n)-1
    while left < right:
        if n[left] != n[right]:
            return False
        left += 1
        right -= 1
    return True
print(palin(n))
```

```python
def sec(n):
    first = float('-inf')
    sec = float('-inf')
    for i in n:
        if i > first:
            sec = first
            first = i
        elif i > sec and i != first:
            sec = i
    return sec

print(sec(n))
```

```python
num = [2,7,11,15]
target = 9

def sum(num,target):
    for i in range(len(num)):
        for j in range(i,len(num)):
            if num[i] + num[j] == target:
                return [i,j]
print(sum(num,target))
```

```python
prices = [7,1,5,3,6,4]

def max_profit(prices):
    min_price = float('inf')
    max_profit = 0

    for price in prices:
        if price < min_price:
            min_price = price
        else:
            profit = price - min_price
            max_profit = max(max_profit, profit)
    return max_profit

print(max_profit(prices))
```

```python
from datetime import datetime

class Recharge:
    def __init__(self, year, month, day, amount=299, validity_days=30):
        self.amount = amount
        self.validity_days = validity_days
        self.recharge_date = datetime(year, month, day)

    def calculate_details(self):
        today = datetime.today()
        total_days = (today - self.recharge_date).days

        if total_days < 0:
            return 0, 0

        # total amount spent till now
        cycles = total_days // self.validity_days + 1
        total_spent = cycles * self.amount

        # balance days in current cycle
        balance_days = self.validity_days - (total_days % self.validity_days)

        # if exactly completed cycle
        if total_days % self.validity_days == 0:
            balance_days = 0

        return balance_days, total_spent

recharge = Recharge(2026, 1, 1)

balance, spent = recharge.calculate_details()

print("Balance Days:", balance)
print("Total Amount Spent:", spent)
```
