```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

n = 20
primes = []
for num in range(2, n + 1):
    if is_prime(num):
        primes.append(num)
print(f'All prime numbers from 1 to {n}: {primes}')
```

```python
s = "A man, a plan, a canal: Panama"
rev=""
sym = [',',':',' ']
cap = ['A','P']
for i in range(len(s)-1,-1,-1):
    if s[i] in sym:
        continue
    elif s[i] in cap:
        rev+=s[i].lower()
    else:
        rev+=s[i]
print(rev)

f=''
for i in s:
    if i in sym:
        continue
    elif i in cap:
        f+=i.lower()
    else:
        f+=i

print(f)

if rev == f:
    print(True)
else:
    print(False)
```

```python
def palindrome(str):
    rev=''
    for i in range(len(str)-1,-1,-1):
        rev+=str[i]
    if rev == str:
        return True
    else:
        return False
print(palindrome('mam'))
print(palindrome('madam'))
print(palindrome('pop'))
print(palindrome('sir'))
```
