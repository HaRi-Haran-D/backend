```python
arr = [5,3,8,1]

for i in range(len(arr)):
    for j in range(0,len(arr)-i-1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1] = arr[j+1], arr[j]
print('Sorted list: ',arr)
```

```python
num = 4567
rev = 0
while num>0:
    rem = num % 10
    rev = (rev * 10) + rem
    num = num // 10
print(rev)
```
