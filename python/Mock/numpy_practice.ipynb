```python
arr = np.array([[1,2,3],[4,5,6]])

print(arr)
#[[1 2 3]
# [4 5 6]]

print(arr.shape)
# rows, columns
#(2, 3)

print(arr.size)
# to find total no. of elements
# 6

print(arr.ndim)
# to find no. of dimensions
# 2

print(arr.dtype)
# to find data type
# int64
```

```python
a = np.array([1,2,3])

# to add any value to the list
print(a + 2)
# [3 4 5]

# to multiply any no. to the list
print(a * 3)
# [3 6 9]

# to find the square root
print(np.sqrt(a))
#[1.         1.41421356 1.73205081]

# to find the sum
print(np.sum(a))
# 6

# to calculate average
print(np.mean(a))
# 2.0
```
